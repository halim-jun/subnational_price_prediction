```python
import xarray as xr
import os
```

```python

for files in os.listdir('../../data/processed/spi/05_spi_final/'):
    if files.endswith('.nc'):
        nc_file_path = '../../data/processed/spi/05_spi_final/' + files
        ds = xr.open_dataset(nc_file_path)
        
        print("Converting to DataFrame...")
        df = ds.to_dataframe()
        df = df.reset_index()
        if not os.path.exists('../../data/processed/spi/06_spi_csv/'): os.makedirs('../../data/processed/spi/06_spi_csv/')
        df.to_csv('../../data/processed/spi/06_spi_csv/' + files.replace('.nc', '.csv'))
```
